```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# G√©n√©rer des donn√©es:
X, y = make_blobs(n_samples=100, centers = 3)
#nb_features = 2 par d√©faut
plt.scatter(X[:,0], X[:, 1])
```

### Entrainer le modele de K-mean Clustering

```python
model = KMeans(n_clusters=3) # 3 centro√Ødes
model.fit(X)
```

### Les centro√Ødes

```python
centres = model.cluster_centers_
```

### Visualiser les Clusters

```python
predictions = model.predict(X)
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(X[:,0], X[:,1], c=predictions)
plt.scatter(centres[0,0] , centres[0,1], c='red')#centro√Øde1
plt.scatter(centres[1,0] , centres[1,1], c='red')#centro√Øde2
plt.scatter(centres[2,0] , centres[2,1], c='red')#centro√Øde3
```
